## Converting machine learning data set into tableau friendly data set

```python
import pandas as pd

hms = pd.read_csv("hms_major_collapse.csv")
```

```python
hms.head()
```

### Transition exercise from encoding

```python
exercise_cat = {
    '1.0' : 'Less than 1 hour',
    '2.0' : '1-2 hours',
    '3.0' : '2-3 hours',
    '4.0' : '3-4 hours',
    '5.0' : '5 or more hours'
}

for index, row in hms.iterrows():
    # reverse the mapping
    for encode, category in exercise_cat.items():
        if str(row['exercise_hours_weekly']) == encode:
            hms.at[index, 'exercise_hours_weekly'] = category 
```

```python
hms.head()
```

### Transition seek help if need column

```python
seek_help_map = {
    0.0: "Disagree",
    1.0: "Partly disagree",
    2.0: "Partly agree",
    3.0: "Agree"
}

hms['seek_help_if_need'] = hms['seek_help_if_need'].map(seek_help_map)
```

```python
hms.head()
```

```python
percneed_map = {
    1.0: "Strongly agree",
    2.0: "Agree",
    3.0: "Somewhat agree",
    4.0: "Somewhat disagree",
    5.0: "Disagree",
    6.0: "Strongly disagree"
}

hms['perceived_help_need'] = hms['perceived_help_need'].map(percneed_map)

know_where_map = {
    1.0: "Strongly agree",
    2.0: "Agree",
    3.0: "Somewhat agree",
    4.0: "Somewhat disagree",
    5.0: "Disagree",
    6.0: "Strongly disagree"
}

hms['know_where_help'] = hms['know_where_help'].map(know_where_map)

diagnosed_map = {
    1.0: "Yes",
    0.0: "No"
}

hms['diagnosed_mental_illness'] = hms['diagnosed_mental_illness'].map(diagnosed_map)


therapy_map = {
    1.0: "Yes",
    0.0: "No"
}

hms['therapy_past12_months'] = hms['therapy_past12_months'].map(therapy_map)

```

```python
hms.info()
```

```python
fincur_map = {
    1.0: "Always stressful",
    2.0: "Often stressful",
    3.0: "Sometimes stressful",
    4.0: "Rarely stressful",
    5.0: "Never stressful"
}

hms['financial_stress'] = hms['financial_stress'].map(fincur_map)


year_map = {
    1.0: "1st year",
    2.0: "2nd year",
    3.0: "3rd year",
    4.0: "4th year",
    5.0: "5th year",
    6.0: "6th year",
    7.0: "7th+ year"
}
hms['year_in_school'] = hms['year_in_school'].map(year_map)

binary_map = {
    1.0: "Yes",
    0.0: "No"
}

hms['suicide_idea'] = hms['suicide_idea'].map(binary_map)
hms['suicide_plan'] = hms['suicide_plan'].map(binary_map)
hms['suicide_attempt'] = hms['suicide_attempt'].map(binary_map)
hms['student_athlete'] = hms['student_athlete'].map(binary_map)
hms['not_white'] = hms['not_white'].map(binary_map)


student_major_map = {
    0: "Arts and Humanities",
    1: "Social Sciences",
    2: "STEM",
    3: "Health and Medicine",
    4: "Business",
    5: "Public and Social Services",
    6: "Education",
    7: "Undecided"
}
hms['student_major'] = hms['student_major'].map(student_major_map)
```

### Transition gender column from encoded values

```python
gender_map = {
    1.0: "Female",
    2.0: "Male"
}

hms['gender_birth'] = hms['gender_birth'].map(gender_map)
```

### Add depression risk labels

```python
hms['depression_risk'] = hms['depression_score'].apply(lambda x: "High Risk" if x >= 10 else "Low Risk")
```
